Criar um Calculador de Descontos

Neste laboratório você escreverá uma função que calcula o preço final de um item após aplicar um desconto percentual.

Por exemplo, se o preço de um item é 50 e um desconto de 20 é aplicado, o valor do desconto é 10 e o preço final é 40.

Objetivo: Cumprir as user stories abaixo para completar o laboratório.

Histórias de Usuário:

1. ✅ Você deve definir uma função chamada apply_discount.

2. ✅ A função apply_discount deve receber exatamente dois parâmetros: price e discount.

3. ✅ Se price não for um número (int ou float), a função deve retornar a string The price should be a number.

4. ✅ Se discount não for um número (int ou float), a função deve retornar a string The discount should be a number.

5. ✅ Se price for menor ou igual a 0, a função deve retornar a string The price should be greater than 0.

6. ✅ Se discount for menor que 0 ou maior que 100, a função deve retornar a string The discount should be between 0 and 100.

7. ✅ Se ambas as entradas forem válidas, a função deve calcular o desconto como uma porcentagem do preço.

8. ✅ A função deve retornar o preço final após aplicar o desconto.

Lista de testes

1.  ✅ Você deve ter uma função chamada apply_discount.
2.  ✅ Sua função apply_discount deve receber dois parâmetros: price e discount.
3.  ✅ Quando apply_discount é chamado com um price (primeiro argumento) que não é um número (int ou float) ele deve retornar The price should be a number.
4.  ✅ Quando apply_discount é chamado com um discount (segundo argumento) que não é um número (int ou float) ele deve retornar The discount should be a number.
5.  ✅ Quando apply_discount é chamado com um price menor ou igual a 0, ele deve retornar The price should be greater than 0.
6.  ✅ Quando apply_discount é chamado com um discount menor que 0 ou maior que 100, ele deve retornar The discount should be between 0 and 100.
7.  ✅ apply_discount(100, 20) deve retornar 80.
8.  ✅ apply_discount(200, 50) deve retornar 100.
9.  ✅ apply_discount(50, 0) deve retornar 50.
10. ✅ Quando apply_discount é chamado com um desconto de 100, ele deve retornar 0.
11. ✅ apply_discount(74.5, 20.0) deve retornar 59.6.


```python
def apply_discount(price, discount):

    if (type(price) != int and type(price) != float):
            return "The price should be a number"

    elif (type(discount) != int and type(discount) != float):
            return "The discount should be a number"
         
    elif price <= 0:
            return "The price should be greater than 0"
            
    elif discount < 0 or discount > 100:
            return "The discount should be between 0 and 100"            
     
    return price-((discount/100)*price)

apply_discount(5, 345)
```
