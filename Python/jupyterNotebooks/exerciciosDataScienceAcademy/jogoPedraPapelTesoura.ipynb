```python
print("""        Jogue
Pedra, Papel e Tesoura
      
      1. Pedra
      2. Papel
      3. tesoura  
      """)

player1 = input('Jogador 1 -> escolha entre "Pedra", "Papel" ou "Tesoura"').lower().strip()
player2 = input('Jogador 2 -> escolha entre "Pedra", "Papel" ou "Tesoura"').lower().strip()

print(f'Jogador 1 escolheu.: {player1}\nJogador 2 escolheu.: {player2}\n')

opcoesValidas = ("pedra", "papel", "tesoura")

if player1 not in opcoesValidas or player2 not in opcoesValidas:
    print(f'Escolha uma opção entre {opcoesValidas[0]},{opcoesValidas[1]} ou {opcoesValidas[-1]}')

elif player1 == player2:
    print("Houve um empate")

elif (player1 == "pedra" and player2 == "tesoura") or (player1 == "tesoura" and player2 == "papel") or (player1 == "papel" and player2 == "pedra"):
    print(f'{player1} vence {player2}, vencedor é Jogador 1!!!')

else:
    print(f'{player2} vence {player1}, vencedor é Jogador 2!!!')


```
